```python
## Import Liberaries
import pandas as pd
import sqlite3
```

```python
## Read the file

df = pd.read_csv('C:\\Users\\subal\\OneDrive\\Desktop\\Documents\\Project\\sql\\Football Players 2015-2024\\Latest Football  Players 2024 Data.csv')
df.head()
```

```python
# import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
df.to_sql("football_data", con =engine)
```

```python
# Select all columns from the football_data table
sql='''
SELECT * FROM football_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select distinct teams from the football_data table
sql='''
SELECT DISTINCT "Teams" FROM football_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Count the number of distinct players 
sql='''
SELECT COUNT(DISTINCT "Players") AS num_players FROM football_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select players who scored more than 20 goals
sql='''
SELECT "Players" FROM football_data
WHERE "Goals" > 20
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select players who scored more than 20 goals
sql='''
SELECT "Players" FROM football_data
WHERE "Goals" > 20
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(20)
```

```python
# Select teams and their total goals
sql='''
SELECT "Teams", SUM("Goals") AS total_goals FROM football_data
GROUP BY "Teams"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the average goals per match for each player
sql='''
SELECT "Players", AVG("Goals" / "Matches") AS avg_goals_per_match FROM football_data
GROUP BY "Players"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the top 5 players with the most assists
sql='''
SELECT "Players", "Assists" FROM football_data
ORDER BY "Assists" DESC
LIMIT 5
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select teams where the average seasons ratings are above 8.0
sql='''
SELECT "Teams" FROM football_data
GROUP BY "Teams"
HAVING AVG(CAST("Seasons Ratings" AS FLOAT)) > 8.0
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select players and their total contributions (goals + assists)
sql='''
SELECT "Players", (SUM("Goals") + SUM("Assists")) AS total_contributions
FROM football_data
GROUP BY "Players"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the average number of matches played per season for each player
sql='''
SELECT "Players", AVG("Matches") AS avg_matches_per_season
FROM football_data
GROUP BY "Players"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select the top 3 teams with the highest total goals in a single season
sql='''
SELECT "Teams", "Seasons", SUM("Goals") AS total_goals
FROM football_data
GROUP BY "Teams", "Seasons"
ORDER BY total_goals DESC
LIMIT 3
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Select players who have played more than 30 matches in a single season
sql='''
SELECT "Players", "Seasons", "Matches"
FROM football_data
WHERE "Matches" > 30
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(8)
```

```python
# Select the number of seasons played by each player
sql='''
SELECT "Players", COUNT("Seasons") AS num_seasons
FROM football_data
GROUP BY "Players"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head(7)
```

```python
# Select the players with the highest goals per match ratio
sql='''
SELECT "Players", (SUM("Goals") / SUM("Matches")) AS goals_per_match
FROM football_data
GROUP BY "Players"
ORDER BY goals_per_match DESC
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql
```

```python
# Calculate the weighted average seasons ratings for each team, weighted by the number of matches played
sql='''
SELECT "Teams", 
SUM(CAST("Seasons Ratings" AS FLOAT) * "Matches") / SUM("Matches") AS weighted_avg_seasons_ratings
FROM football_data
GROUP BY "Teams"
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Rank players based on their total contributions (goals + assists) per season and return the top 10
sql='''
SELECT "Players", "Seasons", (SUM("Goals") + SUM("Assists")) AS total_contributions,
RANK() OVER (PARTITION BY "Seasons" ORDER BY (SUM("Goals") + SUM("Assists")) DESC) AS season_rank
FROM football_data
GROUP BY "Players", "Seasons"
ORDER BY season_rank
LIMIT 10
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Calculate the percentage difference in goals scored between two seasons for each player
sql='''
SELECT "Players", "Seasons",
(LAG("Goals", 1) OVER (PARTITION BY "Players" ORDER BY "Seasons") - "Goals") / LAG("Goals", 1) OVER (PARTITION BY "Players" ORDER BY "Seasons") * 100 AS pct_diff_goals
FROM football_data
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Find the team with the highest average assists per match across all seasons
sql='''
SELECT "Teams", AVG("Assists" / "Matches") AS avg_assists_per_match
FROM football_data
GROUP BY "Teams"
ORDER BY avg_assists_per_match DESC
LIMIT 1
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```

```python
# Select the top 5 players with the highest total goals scored in their career
sql='''
SELECT "Players", SUM("Goals") AS total_goals
FROM football_data
GROUP BY "Players"
ORDER BY total_goals DESC
LIMIT 5
''';

df_sql = pd.read_sql_query(sql, con=engine)
df_sql.head()
```
